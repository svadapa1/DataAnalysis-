##Explotatory Data Analysis using python pandas for a Kaggle Data set

##Importing Libraries

```python
import pandas as pd
import seaborn as sns
```

```python
dataframe = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
dataframe.head(10)
```

```python
dataframe.dtypes
```

```python
dataframe.shape
```

## Let's check out the data of the event which has distance of 50 MIL and 50 KM and held in 2020

```python
dataframe[dataframe['Event distance/length'].isin(['50km','50mi']) & (dataframe['Year of event']== 2020)]
```

## Lets find every event that has been happened in the USA

```python
dataframe[dataframe["Event name"].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

## Let's combine all the filters that we have been used at the top of the note book 

```python
df = dataframe[dataframe['Event distance/length'].isin(['50km','50mi']) & (dataframe['Year of event']== 2020) & (dataframe["Event name"].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
df.head(50)
```

##Let's remove USA from the event name to get accurate information

```python
df['Event name'].str.split('(').str.get(0)
```

```python
df['Event name'] = df['Event name'].str.split('(').str.get(0)
df.head()
```

##Cleaning up the athelete age in the right format

```python
df['Athlete_age'] = 2020 - df['Athlete year of birth']
```

##Removing 'h' from athelete performance time to make it look little accurate

```python
df['Athlete performance'] = df['Athlete performance'].str.split(' ').str.get(0)
```

```python
df.head()
```

## Dropping all the unnessary columns from the data to produce accurate information

```python
df = df.drop(['Athlete year of birth','Athlete age category','Athlete country','Athlete club'],axis = 1)
```

```python
df.isna().sum()
```

```python
df[df['Athlete_age'].isna() == 1]
```

```python
df = df.dropna()
df.shape
```

## Checking for duplicates and resetting the index

```python
df[df.duplicated() == True]
```

```python
df.reset_index(drop=True)
```

## Fixing the date types of each columns

```python
df.dtypes
```

```python
df['Athlete_age'] = df['Athlete_age'].astype(int)
```

```python
df['Athlete average speed'] = df['Athlete average speed'].astype(float)
```

# Renaming column names

```python
df = df.rename(columns={'Event name':'Name of the Event',
                        'Athlete_age':'age',
                        'Event number of finishers':'Finalists',
                        'Year of event':'Year',
                        'Event distance/length':'Distance'})
```

##Charts and Graphs

## Difference between Male and Female for 50 km and 50 mile

```python
df.groupby(['Distance','Athlete gender'])['Athlete average speed'].mean()
```

##What age groups are best in 50mi range (min 20 + Races)

```python
df.query('Distance == "50mi"').groupby('age')['Athlete average speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

##Let's divide each race by month and season

```python
df['Race Month'] = df['Event dates'].str.split(".").str.get(0).astype(int)
```

```python
df.head()
```

```python
df['Race Season'] = df['Race Month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'spring' if x > 2 else 'Winter')
```

```python
df.tail(100)
```

```python
df.groupby('Race Season')['Athlete average speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```

```python
df.query('Distance == "50mi"').groupby('Race Season')['Athlete average speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
